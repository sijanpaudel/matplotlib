```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("Python",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Growth",fontsize=15)
plt.text(2,5,'Sijan',color='k',fontsize=10,style='italic',bbox={"facecolor":"r"})
plt.annotate("java",xy=(2,1),xytext=(3,3),arrowprops=dict(facecolor='k',shrink=100))
plt.legend(["up"],loc=9,facecolor='red',edgecolor='yellow',framealpha=0.5,shadow=True)
plt.show()
```
