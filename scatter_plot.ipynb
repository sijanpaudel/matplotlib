# Scatter Plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color='r')
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors = [10,20,49,30,50,70,39]
sizes = [400,200,400,300,200,100,600]
plt.scatter(day,no,c=colors,s=sizes,alpha=1,marker='*',edgecolor='k',linewidth=2,cmap='viridis')
plt.colorbar()
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors = [10,20,49,30,50,70,39]
sizes = [400,200,400,300,200,100,600]
plt.scatter(day,no,c=colors,s=sizes,edgecolor='k',cmap='BrBG')
t=plt.colorbar()
t.set_label("Color bar",fontsize=20)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,2,4,5,1,6,2]
colors = [10,20,49,30,50,70,39]
sizes = [400,200,400,300,200,100,600]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.5,cmap='viridis')
plt.scatter(day,no2,c="r",s=sizes,alpha=0.5)
t=plt.colorbar()
t.set_label("Color Bar",fontsize=15)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```
