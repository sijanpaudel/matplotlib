# Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.randint(10,60,50)
print(x)
```

```python
no = [22,49, 45, 31, 53, 27, 20, 10, 11, 49, 13, 43, 28, 58, 31, 18, 30, 51, 39, 57, 16, 49, 57, 39,
 15, 23, 41, 54, 19, 44, 49, 33, 32, 16, 57, 35, 50, 19, 52, 43, 15, 33, 24, 18, 21, 20, 24, 58,
 35, 32]

l=[10,20,30,40,50,60]
plt.hist(no,edgecolor='k',bins=l,linewidth=2,color='g',cumulative=True,bottom=10,align='left')
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```
