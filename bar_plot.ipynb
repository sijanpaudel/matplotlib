# Bar Plot

```python
import matplotlib.pyplot as plt
x = ["python","c","c++","java"]
y = [85,70,60,82]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Popularity of Languages",fontsize=15)
c = ['y','b','m','g']
plt.bar(x,y,width=0.4,color=c,align='center',edgecolor='k',linewidth=2,linestyle=':',alpha=1,label="Popularity")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]
width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Popularity of Languages",fontsize=15)
plt.bar(p,y,width, color='r',label='popularity')
plt.bar(p1,z,width,color='b', label='popularity1')
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]
width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]
plt.ylabel("Languages",fontsize=15)
plt.xlabel("No.",fontsize=15)
plt.title("Popularity of Languages",fontsize=15)
plt.barh(p,y,width, color='r',label='popularity')
plt.barh(p1,z,width,color='b', label='popularity1')
plt.yticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
