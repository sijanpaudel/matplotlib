# Box Plot

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50.60,70]

plt.boxplot(x,notch=True,vert=False)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50.60,70]

plt.boxplot(x,widths=0.1,labels=['python'],patch_artist=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50.60,70,120]

plt.boxplot(x,widths=0.1,labels=['python'],showmeans=True,whis=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50.60,70,120]

plt.boxplot(x,widths=0.1,labels=['python'],showmeans=True,sym='g+',boxprops={"color":"b"},capprops=dict(color='y'),
            whiskerprops=dict(color='r'))
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50.60,70]
x1 = [10,20,30,40,50,60,90]

y=[x,x1]
plt.boxplot(y,widths=0.1,labels=['python','jav'],showmeans=True,sym='g+',boxprops={"color":"b"},capprops=dict(color='y'),
            whiskerprops=dict(color='r'))
plt.show()
```
